```python
import pandas as pd

# =========================
# Task 1: Load the Dataset
# =========================

# Direct download link
url = "https://drive.google.com/uc?export=download&id=1Mlu9r3afSGz5_02Y3wDJoQLpXszQXaIU"

# Load dataset with order_id as index
df = pd.read_csv(url, index_col='order_id')

print("First 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())
```

```python
print("\n--- Structural Inspection ---")

# Shape
rows, cols = df.shape
print(f"\nRows: {rows}")
print(f"Columns: {cols}")

# Missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Data types
print("\nData types before conversion:")
print(df.dtypes)

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

print("\nData types after conversion:")
print(df.dtypes)
```

#Task 2 — Structural Inspection

The dataset contains 20 rows and 6 columns.

The following columns have missing values:

quantity → 4 missing values

order_date → 4 missing values

The order_date column was initially stored as an object data type and was converted to datetime using:

pd.to_datetime(df['order_date'])

```python
print("\n--- Statistical Summary ---")

summary = df.describe()
print(summary)

median_price = summary.loc['50%', 'unit_price']
mean_price = summary.loc['mean', 'unit_price']

print(f"\nMedian unit_price: ₹{median_price}")
print(f"Mean unit_price: ₹{mean_price}")

if mean_price > median_price:
    print("unit_price appears to be right-skewed (positively skewed).")
elif mean_price < median_price:
    print("unit_price appears to be left-skewed (negatively skewed).")
else:
    print("unit_price appears symmetric.")
```

#Task 3 — Statistical Summary
The median unit_price is ₹2,500.

The unit_price distribution is right-skewed because:

The mean (₹20,224.60) is significantly greater than the median (₹2,500).

The maximum value (₹85,000) is much larger than the 75th percentile (₹14,625).

This indicates that high-value orders are increasing the average price.
